```python
#import csv
import pandas as pd
data= pd.read_csv('../inputdata/IPL_IMB381IPL2013.csv')
data.head(5)
type(data)
```

```python
#indexing the data
data.iloc[0:5,1:4]
```

```python
#sorting
data[['PLAYER NAME','SOLD PRICE']].sort_values('SOLD PRICE', ascending=False)[0:5]
```

```python
#creating new columns
data['PREMIUM']=data['SOLD PRICE']-data['BASE PRICE']
data[['PLAYER NAME','SOLD PRICE', 'PREMIUM']].sort_values('PREMIUM', ascending=False)[0:5]
```

```python
#grouping and aggregation 
# The groupby operation by default returns
#a data structure called pd dot Series to convert into a dataframe,
#we can call reset index at the end of the expression.

#data.groupby('AGE')['SOLD PRICE'].mean()
data.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
```
