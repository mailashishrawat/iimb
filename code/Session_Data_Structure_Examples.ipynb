```python
weather_df = pd.read_csv('weather.csv')
```

```python
weather_df.head()
```

```python
weather_df.shape
```

```python
weather_formatted = weather_df.melt(id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp')
```

```python
weather_formatted.head()
```

```python
weather_formatted.shape
```

```python
weather_formatted = weather_formatted.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)

weather_formatted = weather_formatted.reset_index()
```
